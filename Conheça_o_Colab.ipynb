```
# ===============================
# Sistema de Cadastro de Alunos
# ===============================

alunos = []  # Lista para armazenar os nomes dos alunos

def cadastrar_aluno():
    nome = input("Digite o nome do aluno para cadastrar: ").strip()
    if nome:
        alunos.append(nome)
        print(f" Aluno '{nome}' cadastrado com sucesso!\n")
    else:
        print(" Nome inválido!\n")

def listar_alunos():
    if alunos:
        print("\n Lista de alunos cadastrados:")
        for i, nome in enumerate(alunos, start=1):
            print(f"{i}. {nome}")
        print()
    else:
        print(" Nenhum aluno cadastrado.\n")

def buscar_aluno():
    nome = input("Digite o nome do aluno para buscar: ").strip()
    if nome in alunos:
        print(f" O aluno '{nome}' está cadastrado.\n")
    else:
        print(f" O aluno '{nome}' não foi encontrado.\n")

def remover_aluno():
    nome = input("Digite o nome do aluno para remover: ").strip()
    if nome in alunos:
        alunos.remove(nome)
        print(f" Aluno '{nome}' removido com sucesso!\n")
    else:
        print(f" O aluno '{nome}' não existe na lista.\n")

def verificar_texto():
    nome = input("Digite o nome para verificar se está na lista: ").strip()
    if nome in alunos:
        print(f" '{nome}' está na lista!\n")
    else:
        print(f" '{nome}' não está na lista.\n")

def comparar_nomes():
    if alunos:
        print(f" Maior nome (ordem alfabética): {max(alunos)}")
        print(f" Menor nome (ordem alfabética): {min(alunos)}\n")
    else:
        print(" A lista está vazia.\n")

def transformar_maiusculas():
    if alunos:
        print(" Nomes em letras maiúsculas:")
        for nome in alunos:
            print(nome.upper())
        print()
    else:
        print(" Nenhum  foi aluno cadastrado.\n")

def remover_ultimo():
    if alunos:
        removido = alunos.pop()
        print(f" Último aluno removido: {removido}\n")
    else:
        print(" A lista não possue nomes.\n")

def encerrar():
    print(" Finalizando o sistema... Até logo!")
    exit()

while True:
    print("=== Sistema de Cadastro de Alunos ===")
    print("1. Cadastrar aluno")
    print("2. Listar alunos")
    print("3. Buscar aluno pelo nome")
    print("4. Remover aluno")
    print("5. Verificar se nome existe (in)")
    print("6. Comparar nomes (maior/menor)")
    print("7. Transformar nomes em maiúsculas")
    print("8. Remover último aluno (pop)")
    print("9. Encerrar sistema")

    opcao = input("Escolha uma dessas opções  (1-9): ")

    if opcao == "1":
        cadastrar_aluno()
    elif opcao == "2":
        listar_alunos()
    elif opcao == "3":
        buscar_aluno()
    elif opcao == "4":
        remover_aluno()
    elif opcao == "5":
        verificar_texto()
    elif opcao == "6":
        comparar_nomes()
    elif opcao == "7":
        transformar_maiusculas()
    elif opcao == "8":
        remover_ultimo()
    elif opcao == "9":
        encerrar()
    else:
        print(" essa Opção e inválida! Tente novamente.\n")
```

```
# 2. Controle de Despesas Pessoais do aluno


despesas = [] #guardar despesas

print('-'*10, 'Controle de despesas', '-'*10)
print('Digite os valores das despesas. Para encerrar, digite 0.')

while True:
  valor = float(input('Digite o valor da despesa: R$ '))
  if valor == 0:
    break
  elif valor < 0:
    print('Valor inválido! Digite um número positivo ou 0 para sair.')
  else:
    despesas.append(valor)



if len(despesas) == 0:
  print('\nNenhuma despesa registrada.')
else:
  total = sum(despesas)
  media = total / len(despesas)
  maior = max(despesas)
  menor = min(despesas)

  print('\n', '-'*10, 'Resumo de Despesas', '-'*10)
  print(f'Total gasto: R$ {total:.2f}')
  print(f'Média de gastos: R$ {media:.2f}')
  print(f'Maior despesa: R$ {maior:.2f}')
  print(f'Menor despesa: R$ {menor:.2f}')
```

```
#. atividade 3

# ============================================
#  CARDÁPIO DA LANCHONETE
# Usando listas paralelas
# ============================================

import os
from time import sleep


def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def linha():
    print("─" * 50)

def titulo(txt):
    linha()
    print(txt.center(50))
    linha()


itens = ["Café ", "Pão pizza", "Sanduíche artesanal", "Suco de uva", "Bolo de prestigio"]
precos = [4.00, 5.00, 9.50, 4.00, 7.00]


def mostrar_cardapio():
    limpar_tela()
    titulo(" CARDÁPIO DA LANCHONETE")
    print(" Nº  | ITENS                     | PREÇO (R$:)")
    linha()
    for i in range(len(itens)):
        print(f" {i+1:<3} | {itens[i]:<24} | {precos[i]:>6.2f}")
    linha()


def sistema_lanchonete():
    pedido = []
    total = 0.0

    while True:
        mostrar_cardapio()
        print("Digite o número do item que você deseja (ou digite 0 para finalizar)\n")

        try:
            opcao = int(input(" Escolha: "))
            if opcao == 0:
                break
            elif 1 <= opcao <= len(itens):
                pedido.append(itens[opcao - 1])
                total += precos[opcao - 1]
                print(f"\n '{itens[opcao - 1]}' adicionado ao pedido!")
            else:
                print("\n essa Opção não existe inválida! Tente novamente.")
        except ValueError:
            print("\n Digite apenas números válidos.")

        sleep(1.5)

    mostrar_resumo(pedido, total)


def mostrar_resumo(pedido, total):
    limpar_tela()
    titulo(" RESUMO DO PEDIDO")

    if pedido:
        print("Itens escolhidos:\n")
        for item in pedido:
            print(f"  • {item}")
        linha()
        print(f" Total a pagar: R$ {total:.2f}")
    else:
        print("Nenhum item foi selecionado.")

    print("\nObrigado pela preferência! \n")
    input("Pressione ENTER para sair...")
    print("atendimento encerrado!")
# --------------------------------------------
# Execução do sistema
# --------------------------------------------
sistema_lanchonete()
```

```
# 4. Controle de Estoque da Lanchonete

estoque = {} #armazenamento

def adicionar_produto():
  nome = input('Nome do produto: ').strip().capitalize()
  if nome in estoque:
    print('Produto já cadastrado. Use a opção de atualização.')
  else:
    qtd = int(input('Quantidade inicial: '))
    estoque[nome] = qtd
    print(f'Produto {nome} atualizada para {qtd} unidades.')


  def atualizar_produto():
    nome = input('Nome do produto a atualizar: ').strip().capitalize()
    if nome in estoque:
      qtd = int(input('Nova quantidade: '))
      estoque[nome] = qtd
      print(f'Quantidade de {nome} atualizada para {qtd}.')
    else:
      print('Produto não encontrado.')


  def listar_produtos():
    if not estoque:
      print('Nenhum produto cadastrado.')
    else:
      print('='*10,'Lista de Produtos', '='*10)
      for nome, qtd in estoque.items():
        print(f'{nome}: {qtd} unidade(s)')



  def mostrar_zerados():
    zerados = [nome for nome, qtd in estoque.items() if qtd == 0]
    if not zerados:
      print('Nenhum produto com estoque zerado.')
    else:
      print('='*10,'Produtos com estoque zerado', '='*10)
      for nome in zerados:
        print(nome)



  # Menu Principal

  while True:
    print('-'*10, 'Controle de Estoque', '-'*10)
    print('1. Adicionar Produto')
    print('2. Atualizar Produto')
    print('3. Listar Produtos')
    print('4. Mostrar Produtos com Estoque Zerado')
    print('5. Sair')

    opcao = input('Escolha uma opção: ')

    if opcao == '1':
      adicionar_produto()
    elif opcao == '2':
      atualizar_produto()
    elif opcao == '3':
      listar_produtos()
    elif opcao == '4':
      mostrar_zerados()
    elif opcao == '0':
      print('Encerrando o programa...')
      break
    else:
      print("Opção inválida! Tente novamente.")
```
